# L'analyse en composantes principales avec scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

#charger un jeu de données
iris = datasets.load_iris()
```

```python
#définir les variables
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
#print PCA.__doc__
```

```python
#lancer l'ACP
acp = PCA(n_components=2)
```

```python
#on crée un array avec les projection des X sur les deux facteurs
X_r = acp.fit(X).transform(X)
```

```python
X_r.shape
```

```python
# affichage de la variance expliquée
print(str(acp.explained_variance_ratio_))
```

```python
#projection des individus
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('ACP sur les iris')
plt.xlabel('Facteur1')
plt.ylabel('Facteur2')
```

```python
#utilisation de matplotlib pour l'ACP
from matplotlib.mlab import PCA as mlabPCA

#lancement du modèle
mlab_pca = mlabPCA(iris.data)
```

```python
#représentation des points
plt.plot(mlab_pca.Y[0:50,0],mlab_pca.Y[0:50,1],
         'o', markersize=7, color='blue', alpha=0.5, label='setosa')
plt.plot(mlab_pca.Y[50:100,0], mlab_pca.Y[50:100,1],
         '^', markersize=7, color='red', alpha=0.5, label='versicolor')
plt.plot(mlab_pca.Y[100:150,0], mlab_pca.Y[100:150,1],
         '*', markersize=7, color='green', alpha=0.5, label='virginica')

plt.xlabel('Facteur 1')
plt.ylabel('Facteur 2')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.legend()
plt.title('Representation des deux axes factoriels')
```
