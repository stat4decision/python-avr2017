A partir d’un jeu de données, représentez un nuage de points basé sur deux variables continues


```python
frame_house=pd.read_csv("housing.csv")
```

```python
frame_house.mean()
```

```python
plt.scatter("longitude","latitude",s=frame_house["population"]/100,
            c="median_house_value",alpha=0.1,data=frame_house,label="population")
plt.colorbar()
plt.legend()
```

Ajout de la carte de la Calif

```python
import matplotlib.image as mpimg
calif_image=mpimg.imread("California.png")

plt.scatter("longitude","latitude",s=frame_house["population"]/100,
            c="median_house_value",alpha=0.4,data=frame_house,label="population",cmap=plt.get_cmap("rainbow"))
plt.colorbar()
plt.imshow(calif_image, alpha=0.5, extent=[-124.55, -113.80, 32.45, 42.05])
plt.legend()
```
