# Les k-means pour traiter une image

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

On récupère les données à partir de l'image

```python
from scipy import misc
face = misc.imread('New York.jpg')

plt.imshow(face)
plt.grid(False)
```

```python
face.shape
```

```python
#on réduit le nombre de pixels
image=face[::3,::3]
```

```python
plt.imshow(image)
```

```python
#on passe de 3 dimensions à 2 dimensions
x=(image/255.0).reshape(-1,3)
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
#on crée le modèle
modele=KMeans(n_clusters=16)
```

```python
#on obtient les classes à partir des données de l'image (16 classes)
labels=modele.fit_predict(x)
```

```python
#on obitent les centroïdes de chaque classe
colors=modele.cluster_centers_
```

```python
print(colors.shape, labels.shape)
labels
```

```python
#reconstruire l'image
new_image= colors[labels]
```

```python
#passage à 3 dimensions
new_image=new_image.reshape(image.shape)
```

```python
#changement de format
new_image=(new_image*255.0).astype(np.uint8)
```

```python
plt.imshow(new_image)
```
